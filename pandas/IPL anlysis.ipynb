```python
def run_scored(batsman_name):
    
    # Get the delivery data for the batsman.
    vk = delivery[delivery['batsman'] == batsman_name]
    
    # Group the data by bowling team and sum the batsman runs.
    runs_by_team = vk.groupby('bowling_team')['batsman_runs'].sum()

    # Sort the data by descending order of runs scored.
    team = runs_by_team.sort_values(ascending=False).head(1).index[0]
    print(team)
```

```python
run_scored('DA Warner')
```

```python

#Find most destructive batsman in death overs. Overs between 16-20!
death_overs=delivery[delivery['over']>15]
```

```python
all_batsman=death_overs.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
batsman_list=all_batsman[x].index.tolist()
```

```python
final=delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs=final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()

print(runs)
print(balls)
```

```python
sr = (runs/balls)*100
```

```python
sr
```

```python
sr.sort_values()
```
