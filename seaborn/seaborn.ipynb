```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('tips.csv')
```

```python
data.head()
```

```python
#Rel plot between total_bill and tip
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data)  
```

```python
#the hue parameter
sns.relplot(x='total_bill',y='tip',kind='scatter',hue='smoker',data=data)
```

```python
#the style parameter
sns.relplot(x='total_bill',y='tip',kind='scatter',hue='smoker', style='sex',data=data)
```

```python
#size parameter
sns.relplot(x='total_bill',y='tip',hue='smoker', style='sex',size='size',sizes=[15,200],data=data)
```

```python
#same thing can be done with scatterplot as well as shown below:-
sns.scatterplot(x='total_bill',y='tip',hue='smoker', style='sex',data=data)
```
